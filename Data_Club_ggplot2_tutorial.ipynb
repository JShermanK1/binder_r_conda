```r
#Make a scatter plot
plot1 = ggplot(data = DNase, mapping = aes(x = conc, y = density, colour = as.factor(Run)))+
        geom_point()+
        labs(color = "Run")+
        facet_wrap(~ Run)+
        ggtitle("DNAse Results")+
        xlab("Concentration")+
        ylab("Density")
        
plot1
```
